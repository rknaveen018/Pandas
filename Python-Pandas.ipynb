## Grouping and Aggregation

```python
df.groupby('gender').agg({'tenure':'mean'})
```

**Find the avg monthly charges and the count of customers per contract type but consider only churn customers**

```python
df[df['Churn'] == 'Yes'].groupby("Contract").agg({'MonthlyCharges':'mean', 'customerID':'count'})
```

```python
df.head(2)
```

```python
df.groupby(['gender', 'Churn']).agg({'customerID':'count'})
```

```python
df.pivot_table(index = 'gender', columns = 'Churn', values = 'customerID', aggfunc = 'count')
```

**Find the average tenure for churn and non-churn customers under various paymentmethods**

```python
df.pivot_table(index = 'Churn', columns = 'PaymentMethod', values = 'tenure', aggfunc = 'mean')
```
